```python
from ultralytics import YOLO

# Load the pre-trained YOLOv8 model
model = YOLO("yolov8m.pt")

# Train the model
model.train(data=f"/content/drive_Posefly/MyDrive/Posefly/data.yaml", epochs=50, imgsz=640, batch=16, workers=4)
```

```python
model.save("/content/drive_Posefly/MyDrive/Posefly.pt")
```

```python
#uploading and testing the model with an image

from google.colab import files  #imports the colab module to upload the files from local system

# Upload a test image
uploaded = files.upload()  #Opens chooser for image(.jpg,.png)

# Run inference
results = model(list(uploaded.keys())[0], max_det=3)  #grabs the name of the uploaded file and results contain the detection bounding box, class labels and confidence scores
#results = model(list(uploaded.keys())[0], max_det=2, conf=0.5,iou=0.4)
#conf = confidence rate; iou = Intersection-over-union ratio

results[0].show()  # Display detections over the image with box,label,score etc.
```
